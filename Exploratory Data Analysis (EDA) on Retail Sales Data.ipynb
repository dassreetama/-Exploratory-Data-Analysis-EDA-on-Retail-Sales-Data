```python
import pandas as pd

# Load the dataset
df1 = pd.read_csv(r"C:\Users\Sreetama\OneDrive\Desktop\oasis project\retail_sales_dataset.csv")
df2 = pd.read_csv(r"C:\Users\Sreetama\OneDrive\Desktop\oasis project\menu.csv")

# Print all column names
print("Columns in df1:", df1.columns)
print("Columns in df2:", df2.columns)
```

```python
print(df1.head())
```

```python
# Calculate descriptive statistics using the correct column names
mean_sales = df1['Total Amount'].mean()
median_sales = df1['Total Amount'].median()
mode_sales = df1['Total Amount'].mode()[0]
std_sales = df1['Total Amount'].std()

print(f'Mean Sales: {mean_sales}')
print(f'Median Sales: {median_sales}')
print(f'Mode Sales: {mode_sales}')
print(f'Standard Deviation of Sales: {std_sales}')
```

```python
# Calculate descriptive statistics using the correct column names
mean_calories = df2['Calories'].mean()
median_calories = df2['Calories'].median()
mode_calories = df2['Calories'].mode()[0]
std_calories = df2['Calories'].std()

print(f'Mean Calories: {mean_calories}')
print(f'Median Calories: {median_calories}')
print(f'Mode Calories: {mode_calories}')
print(f'Standard Deviation of Calories: {std_calories}')
```

```python
import matplotlib.pyplot as plt

# Convert 'Date' column to datetime
df1['Date'] = pd.to_datetime(df1['Date'])

# Group by date and sum the total sales for each day
daily_sales = df1.groupby('Date')['Total Amount'].sum()

# Plot sales over time
plt.figure(figsize=(12, 6))
daily_sales.plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.show()
```

```python
# Customer Analysis
# Analyze customer demographics and purchasing behavior

# Group by Customer ID and calculate total amount spent by each customer
customer_spending = df1.groupby('Customer ID')['Total Amount'].sum()

# Calculate average spending per customer
average_spending_per_customer = customer_spending.mean()

# Print top spending customers
top_customers = customer_spending.nlargest(5)  # Get top 5 spending customers
print("Top Spending Customers:")
print(top_customers)

# Calculate average age of customers
average_age = df1['Age'].mean()
print("Average Age of Customers:", average_age)

# Calculate total sales per product category
sales_per_category = df1.groupby('Product Category')['Total Amount'].sum()

# Print top selling product categories
print("Top Selling Product Categories:")
print(sales_per_category.nlargest(5))

```

```python
# Product Analysis
# Analyze top-selling products

# Group by product and calculate total sales
product_sales = df1.groupby('Product Category')['Total Amount'].sum()

# Print top selling products
print("Top Selling Products:")
print(product_sales.nlargest(5))
```
